```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("House_Rent_Dataset.csv")
```

```python
data
```

```python
# City-Average Rent Price graph
cities = data['City']
rent_prices = data.groupby('City')['Rent'].mean().reset_index()

plt.figure(figsize=(10,6))
plt.bar(rent_prices['City'], rent_prices['Rent'], color='skyblue')
plt.xlabel("City")
plt.ylabel("Average Rent Price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Analyzing the relationship between cities and BHK amount
cities = data['City']
rent_prices = data.groupby('City')['BHK'].mean().reset_index()

plt.figure(figsize=(10,6))
plt.bar(rent_prices['City'], rent_prices['BHK'], color='skyblue')
plt.xlabel("City")
plt.ylabel("Average BHK Amount")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Analyzing the relationship between cities and apartment sizes
cities = data['City']
rent_prices = data.groupby('City')['Size'].mean().reset_index()

plt.figure(figsize=(10,6))
plt.bar(rent_prices['City'], rent_prices['Size'], color='skyblue')
plt.xlabel("City")
plt.ylabel("Apartment Size")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Analyzing the relationship between cities and bathrooms
cities = data['City']
rent_prices = data.groupby('City')['Bathroom'].mean().reset_index()

plt.figure(figsize=(10,6))
plt.bar(rent_prices['City'], rent_prices['Bathroom'], color='skyblue')
plt.xlabel("City")
plt.ylabel("Bathroom")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
